# 6. Step-by-step hierarchical clustering
---

The general idea behind hierarchical clustering with the agglomeration method is to start with each data point as its own cluster and then repeatedly merge the closest clusters into larger clusters. Here's how to do it:

1) Calculate the pairwise distances between all data points.

2) Find the two closest clusters and merge them into a single cluster.

3) Update the distance matrix to reflect this merge.

4) Repeat steps 2-3 until we have the desired number of clusters.

```python
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import pdist, squareform
import numpy as np
import pandas as pd

# Selecting a subset of the data and focusing on numerical variables for clustering
data_subset = cleaned_data[['Global_Sales', 'Critic_Score', 'Sales_Per_User', 'Critic_User_Score_Ratio']]

# 1. Preprocess the Data: Normalize the data
scaler = StandardScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data_subset), columns=data_subset.columns)

# Display the first few rows of the scaled data
data_scaled.head()
```

```python
from scipy.cluster.hierarchy import linkage, fcluster

# Step-by-step agglomerative hierarchical clustering
def hierarchical_clustering(df, max_clusters):
    # Perform hierarchical clustering
    Z = linkage(df, method='ward')

    # Initialize the current number of clusters
    current_clusters = df.shape[0]

    # Loop to merge clusters step-by-step
    for i, row in enumerate(Z):
        # Update the current number of clusters
        current_clusters -= 1

        # Print update
        print(f"Step {i+1}: Merged clusters {int(row[0])} and {int(row[1])}. Total clusters: {current_clusters}.")

        # Stop if we reach the desired number of clusters
        if current_clusters == max_clusters:
            print("Reached the target number of clusters.")
            break

    # Create cluster labels
    labels = fcluster(Z, max_clusters, criterion='maxclust')

    return labels

# Run hierarchical clustering and get labels
max_clusters = 5
labels = hierarchical_clustering(data_scaled, max_clusters=max_clusters)

# Show first few labels
labels[:10]

```

# 8. Step-by-step K-means
---

```python
def kmeans(data, k, max_iters=100, tolerance=1e-4):
    """
    K-means clustering algorithm.

    Parameters:
    data (pd.DataFrame): The input data.
    k (int): The number of clusters.
    max_iters (int): Maximum number of iterations.
    tolerance (float): Tolerance to declare convergence.

    Returns:
    np.array: Cluster labels for each data point.
    """
    # 1. Select Initial Centroids
    centroids = data.sample(n=k).values

    for iteration in range(max_iters):
        # 2. Assign Clusters
        distances = np.linalg.norm(data.values - centroids[:, np.newaxis], axis=2)
        cluster_labels = np.argmin(distances, axis=0)

        # 3. Update Centroids
        new_centroids = np.array([data.values[cluster_labels == z].mean(axis=0) for z in range(k)])

        # Check for convergence (if centroids do not change)
        if np.all(np.abs(new_centroids - centroids) < tolerance):
            break

        centroids = new_centroids

    return cluster_labels

# Apply K-means clustering
k = 4
cluster_labels_kmeans = kmeans(data_scaled, k)

# Display the first few cluster labels
cluster_labels_kmeans[:20]
```

```python
from sklearn.decomposition import PCA
import seaborn as sns

# Reduce data to two dimensions for visualization
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Create a scatter plot of K-means clustering results
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_pca[:, 0], y=data_pca[:, 1], hue=cluster_labels_kmeans, palette='viridis', legend='full')
plt.title('K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

Basically, the 'kmeans' function performs K-means clustering on the given DataFrame 'df' with k clusters.

It initializes the centroids randomly from the data points in df. It then iteratively updates the centroids until they no longer change.

After running the kmeans function, the DataFrame df_for_clustering will have an additional column called 'Cluster' that contains the cluster number (from 0 to k  1) for each data point.

# 10. kNN Function
---

```python
from collections import Counter

def predictKNN(k, data_point, data, labels):
    """
    k-Nearest Neighbors classification.

    Parameters:
    k (int): The number of nearest neighbors to consider.
    data_point (np.array): A vector of r numbers representing the data point to be classified.
    data (pd.DataFrame): A dataframe with r columns containing the feature values.
    labels (np.array): A vector of labels for classification.

    Returns:
    int: The predicted classification for the data point.
    """
    # Calculate Euclidean distances between the data point and all points in the dataset
    distances = np.linalg.norm(data.values - data_point, axis=1)

    # Get indices of the k nearest neighbors
    k_neighbors_indices = np.argpartition(distances, k)[:k]

    # Get labels of the k nearest neighbors
    k_neighbor_labels = labels[k_neighbors_indices]

    # Return the most common label among the k neighbors
    return Counter(k_neighbor_labels).most_common(1)[0][0]

# Generate random data points and run the predictKNN function
np.random.seed(42)
results = []
for _ in range(5):
    random_data_point = np.random.rand(data_scaled.shape[1])
    predicted_label = predictKNN(k=3, data_point=random_data_point, data=data_scaled, labels=cluster_labels_sklearn)
    results.append((random_data_point, predicted_label))

# Display the results
results
```
